```python
import matplotlib.pyplot as plt
import numpy as np
plt.suptitle("Graph")
x=np.array([1,2])
y=np.array([4,5])
plt.subplot(3,3,1)

plt.plot(x,y,'o')
plt.title("Tile 1")
plt.xlabel(x,color='blue')
plt.ylabel(y, color='blue')


x1=np.array([0,1,2,3])
y1=np.array([3,8,1,10])

x2=np.array([0,1,2,3])
y2=np.array([8,4,6,3])
plt.subplot(3,3,2)
plt.plot(x1,y1,marker='*',lw="2.5")
plt.plot(x2,y2,linestyle='dotted')


x3 = np.array([0, 1, 2, 3])
y3 = np.array([3, 8, 1, 10])
plt.subplot(3,3,3)
plt.plot(x3,y3,marker='o',ms=10,mfc='yellow',mec='black')

x4 = np.array([2,5])
y4 = np.array([2,8])

plt.subplot(3,3,4)
plt.grid()
plt.plot(x4,y4)


    
x5 = np.array([5,7,8,7,2])
y5 = np.array([99,86,87,88,111]) 
plt.subplot(3,3,5) 
plt.scatter(x5,y5)
x6 = np.array([6,7,10,3,8])
y6 = np.array([55,87,88,83,101])
plt.subplot(3,3,5) 
plt.scatter(x6, y6,color='r')

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
plt.subplot(3,3,6)
plt.bar(x,y,color='r')

plt.show()

```

```python

import matplotlib.pyplot as plt


sizes = [35, 20, 20, 25]
labels = ['Category A', 'Category B', 'Category C', 'Category D']
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']
explode = (0.1, 0, 0, 0)


plt.figure(figsize=(6, 6))
plt.pie(
    sizes,
    labels=labels,
    colors=colors,
    explode=explode,
    startangle=-90,
    shadow=True,
    autopct='%1.1f%%'
)
plt.axis('equal')  
plt.legend(labels, title='Categories', loc='upper right')

plt.show()
```
